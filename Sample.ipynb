```python
import pandas as pd
sum_of_rows = df.sum(axis=1)
sum_of_columns = df.sum(axis=0)
mean_of_rows = df.mean(axis=1)
mean_of_columns = df.mean(axis=0)
variance_of_rows = df.var(axis=1)
variance_of_columns = df.var(axis=0)
results_df = pd.DataFrame({
    'Sum of Rows': sum_of_rows,
    'Sum of Columns': sum_of_columns,
    'Mean of Rows': mean_of_rows,
    'Mean of Columns': mean_of_columns,
    'Variance of Rows': variance_of_rows,
    'Variance of Columns': variance_of_columns
})
print(results_df.to_string())
```

```python
!pip install scipy
import scipy as sp
from scipy.stats import f as scipysf
grand_mean = df.values.mean()
row_means = df.mean(axis=1)

SSB = df.shape[1] * sum((row_means - grand_mean)**2) #sum of square between rows
deviations = df.sub(row_means,axis=0)
SSW = (deviations**2).sum().sum()   #sum of square within rows
df_rows = df.shape[0] - 1                            #degrees of freedom for rows
MSB = SSB / df_rows                                   #mean square between rows
df_within_rows = (df.shape[0]-1) * (df.shape[1] - 1)   #degree of freedom within rows
MSW = SSW / df_within_rows                            #mean square within rows
F_rows = MSB / MSE                                   #F value for rows
p_value_rows = 1 - scipysf.cdf(F_rows, df_rows, df_within_rows)   #p value of rows
column_means = df.mean(axis=0)                                    #column mean
SSC = df.shape[0] * sum((column_means - grand_mean)**2)            #sum of square of columns
df_columns = df.shape[1] - 1                                        #degrees of freedom for columns
MSC = SSC / df_columns                                               #mean square of columns
SSE=SSW-SSC
MSE = SSE / df_within_rows
F_columns = MSC / MSE                                                #F value for columns

p_value_columns = 1 - scipysf.cdf(F_columns, df_columns, df_within_rows)

print("Sum of squares between rows (SSB):", SSB)
print("Sum of squares within rows (SSW):", SSW)
print("Degrees of freedom for rows:", df_rows)
print("Mean squares between rows (MSB):", MSB)
print("Mean squares within rows (MSW):", MSE)
print("F value of rows:", F_rows)
print("P value of rows:", p_value_rows)
print("Sum of squares of columns (SSC):", SSC)
print("Degrees of freedom for columns:", df_columns)
print("Mean squares of columns (MSC):", MSC)
print("F value of columns:", F_columns)
print("P value of columns:", p_value_columns)
```

```python
print("2-way ANOVA Results")
print()
print("Source of Variation | Sum of Squares | Degrees of Freedom | Mean Square | F Value | P Value")
print("------------------ | --------------- | -----------------  | ----------- | --------| --------")
print(f"Rows             | {SSB:.4f}            | {df_rows}              | {MSB:.4f}      | {F_rows:.4f}  | {p_value_rows:.4f}")
print(f"Columns          | {SSC:.4f}            | {df_columns}           | {MSC:.4f}      | {F_columns:.4f} | {p_value_columns:.4f}")
print(f"Within Rows      | {SSW:.4f}            | {df_within_rows}       | {MSW:.4f}      |          |")
print(f"Error            | {SSE:.4f}            | {df_within_rows}       | {MSE:.4f}")
```
